```python
import pandas as pd
```

```python
admission_df = pd.read_csv('ADMISSIONS.csv')
admission_df = admission_df[['subject_id', 'admittime', 'dischtime', 'deathtime', 'admission_type', 'ethnicity', 'diagnosis', 'hospital_expire_flag']]
```

```python
inputevents_mv_df = pd.read_csv('INPUTEVENTS_MV.csv')
inputevents_mv_df = inputevents_mv_df[['subject_id', 'starttime', 'endtime', 'itemid', 'amount', 'amountuom', 'rate', 'rateuom', 'patientweight']]
```

```python
admission_df
```

```python
inputevents_mv_df
```

```python
admission_df['visit'] = admission_df.groupby('subject_id').cumcount() + 1
```

```python
admission1_df = admission_df.sort_values(by=['subject_id', 'admittime'])
inputevents_mv1_df = inputevents_mv_df.sort_values(by=['subject_id', 'starttime'])

inputevents_mv1_df['visit'] = 0

for index, row in inputevents_mv1_df.iterrows():
    subject_id = row['subject_id']
    starttime = row['starttime']
    endtime = row['endtime']

    # Find the corresponding admission for the subject
    matching_admission = admission1_df[(admission1_df['subject_id'] == subject_id) & (admission1_df['admittime'] <= starttime) & (admission1_df['dischtime'] >= endtime)]
    if not matching_admission.empty:
        # If a matching admission is found, mark the visit based on its index
        inputevents_mv1_df.at[index, 'visit'] = matching_admission['visit']
```

```python
unique_itemids = inputevents_mv1_df['itemid'].unique()

# Create a list of unique SUBJECT_ID + visit combinations
unique_subject_visits = inputevents_mv1_df['subject_id'].astype(str) + inputevents_mv1_df['visit'].astype(str)
unique_subject_visits = unique_subject_visits.unique()

# Initialize an empty DataFrame with rows and columns
matrix_df = pd.DataFrame(0, index=unique_subject_visits, columns=unique_itemids)

# Iterate through inputevents_mv_df to fill the matrix
for index, row in inputevents_mv1_df.iterrows():
    subject_visit = str(row['subject_id']) + str(row['visit'])
    item_id = row['itemid']
    matrix_df.at[subject_visit, item_id] = 1
```
